Exceptions are errors detected during execution. Let's learn how to handle them gracefully!

## ðŸ”¹ 1. Basic `try` and `except`

```python
x = 10 / 0
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")
```

## ðŸ”¹ 2. Handling Multiple Exceptions

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

## ðŸ”¹ 3. Using `else` and `finally`

```python
try:
    num = 0
    result = 10 / num
except ZeroDivisionError:
    print("Error: Division by zero.")
else:
    print(f"Success! Result is {result}.")
finally:
    print("This will run no matter what.")
```

## ðŸ”¹ 4. Raising Exceptions Manually

```python
def withdraw(amount):
    balance = 100
    if amount > balance:
        raise ValueError("Insufficient funds!")
    else:
        print("Withdrawal successful!")

try:
    withdraw(150)
except ValueError as e:
    print(e)
```

## ðŸ”¹ 5. Custom Exception Class

```python
class NegativeNumberError(Exception):
    pass

def square_root(num):
    if num < 0:
        raise NegativeNumberError("Cannot take square root of a negative number.")
    return num ** 0.5

try:
    print(square_root(-9))
except NegativeNumberError as e:
    print(e)
```
