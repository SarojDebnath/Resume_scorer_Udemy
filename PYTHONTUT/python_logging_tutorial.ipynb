## ðŸ”¹ 1. Basic Logging Example

```python
import logging

# Configure basic logging
logging.basicConfig(level=logging.INFO)

logging.info("This is an info message.")
logging.warning("This is a warning message.")
logging.error("This is an error message.")
```

## ðŸ”¹ 2. Logging to a File

```python
import logging

logging.basicConfig(filename='C:/Users/sarojd/udemy/HandsOnRobotics/python_basics/app.log',
                    level=logging.DEBUG,
                    format='%(asctime)s - %(levelname)s - %(message)s')

logging.debug("Debug message: Useful for developers.")
logging.info("Info message: Program is running smoothly.")
logging.warning("Warning message: Something might be wrong.")
logging.error("Error message: An error occurred.")
```

## ðŸ”¹ 3. Using Different Log Levels

```python
for level in [logging.DEBUG, logging.INFO, logging.WARNING, logging.ERROR, logging.CRITICAL]:
    logging.log(level, f"This is a {logging.getLevelName(level)} message.")
```

## ðŸ”¹ 4. Example: Logging in a Function

```python
def divide(a, b):
    logging.info(f"Dividing {a} by {b}")
    try:
        result = a / b
        logging.debug(f"Result is {result}")
        return result
    except ZeroDivisionError:
        logging.error("Attempted to divide by zero!")
        return None

divide(10, 2)
divide(10, 0)
```
